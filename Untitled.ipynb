```python

States=vaccine['State'].unique()
States
```

```python

State_wise_data=pd.DataFrame({}) #Creating a empty Dataframe To store state_wise _data.
```

```python
for i in range(len(States)): #adding state data to DataFrame
    State_wise_data=State_wise_data.append(list(vaccine[vaccine['State']==States[i]][-9:-8].values))
```

```python
State_wise_data[-37:]
```

```python
col=vaccine.columns 
```

```python
State_wise_data.columns=col 
data1=State_wise_data[-37:]
data1.sort_values(by='Total Individuals Vaccinated',ascending=False)
```

```python
data1.set_index(np.arange(len(data1)),inplace=True) #Setting up index
```

```python
covisheild=data1['Total CoviShield Administered'].sum()
covaxin=data1['Total Covaxin Administered'].sum()
sputnik=data1['Total Sputnik V Administered'].sum()
print("Total CovidShield Vaccinated:",covisheild,"\nTotal Covaxin vaccinated:",covaxin,"\nTotal Sputnik vaccinated:",sputnik)
total=(data1['First Dose Administered'].sum()+data1['Second Dose Administered'].sum())
print("Percentage of Covaxin Vaccine Administered","{:.2f}".format((covaxin/(covaxin+covisheild+sputnik))*100),"%")
print("Percentage of CoviSheild Vaccine Administered","{:.2f}".format((covisheild/(covaxin+covisheild+sputnik))*100),"%")
print("Percentage of Sputnik V  Vaccine Administered","{:.2f}".format((sputnik/(covaxin+covisheild+sputnik))*100),"%")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.style.use('ggplot')
labels='CoviShield','Covaxin'
ax.pie([covisheild,covaxin],autopct ='% 1.01f %%',startangle=90,labels=labels)
ax.axis('equal')
plt.title('Percentage of Each Vaccine Administered  ')
plt.legend(facecolor='pink')
plt.show()
```
